# Aufgaben zu:Klassifikation von Merkmalen

## 1. Aufgabe

Bestimme die Art der Skala (Nominal, Ordinal, Kardinal) bei der Beurteilung der folgenden Merkmale:  
1. Anzahl der Insassen in einem PKW bei der Verkehrszählung  
2. Reisegeschwindigkeit bei Flugzeugen  
3. Schultypen  
4. Temperaturangaben in 0C  
5. Zugriffszeiten auf Daten beim PC  
6. Fassungsvermögen von Binnenschiffen  
7. Bewertung beim Eiskunstlauf  
8. Ölverbrauch in einem Einfamilienhaus pro Jahr  
9. Stärke von Erdbeben  
10. Einteilung von Schülern nach ihrer Nationalität  
12. Einteilung von Bediensteten einer Firma nach ihrem Bruttogehalt  
13. Sehstärke in Dioptrien  
14. Intelligenzquotient  
15. Einteilung der Schüler des Berufskollegs nach Wohnort

### Lösung

1. **Metrische Skala (Kardinalskala):**
Da die Anzahl der Insassen in einem PKW durch Zählen ermittelt wird, handelt es sich um ein quantitatives Merkmal. Da die Werte abzählbar sind, ist dies eine Kardinalskala, genauer eine diskrete Skala.

2. **Metrische Skala (Kardinalskala)**

3. **Nominalskala:**
Schultypen sind kategoriale Daten ohne ordinale Reihenfolge, was sie zu einem nominal skalierten Merkmal macht.

4. **Metrische Skala (Kardinalskala):**
Temperaturen in Grad Celsius sind kontinuierlich, jedoch ist Celsius eine Intervallskala, nicht eine Kardinalskala. Das könnte zu einer Diskussion führen, jedoch wird oft die gesamte Kardinalskala als Oberkategorie betrachtet, daher ist es in diesem Kontext akzeptabel.

5. **Metrische Skala (Kardinalskala):**
Zugriffszeiten sind kontinuierlich und können daher korrekt als Kardinalskala klassifiziert werden.

6. **Metrische Skala (Kardinalskala)**

7. **Ordinalskala:**
Die Bewertung beim Eiskunstlauf erfolgt in geordneten Kategorien (z.B. Punkte), die als ordinal klassifiziert werden, da die Abstände nicht zwingend gleich sind.

8. **Metrische Skala (Kardinalskala)**

9. **Kardinalskala:**
Die Stärke von Erdbeben wird auf der Richterskala gemessen, was eine logarithmische Skala ist.

10. **Nominalskala:**
Nationalität ist ein kategoriales Merkmal ohne natürliche Reihenfolge, daher nominal.

11. **Kardinalskala**

12. **Kardinalskala**
Sehstärke in Dioptrien ist eine kontinuierliche quantitative Größe.

13. **Intervallskala:**
Der IQ wird auf einer Intervallskala gemessen, da es keinen absoluten Nullpunkt gibt, jedoch gleichwertige Abstände zwischen den Werten bestehen.

14. **Nominalskala:**
Wohnort ist ein kategoriales Merkmal ohne ordinale Reihenfolge, daher nominal.

## 2. Aufgabe

Welche quantitativen Merkmale aus Aufgabe 1 haben eine diskrete und welche haben eine stetige metrische Skala?

### Lösung

Ergebnisse:  
Zu betrachten sind nur die Merkmale mit metrischer Skala.  

1. Anzahl der Insassen in einem PKW bei der Verkehrszählung: Metrische Skala ist diskret.  
2. Reisegeschwindigkeit bei Flugzeugen: Metrische Skala ist stetig.  
3. Temperaturangaben in 0C: Metrische Skala ist stetig.  
4. Zugriffszeiten auf Daten beim PC: Metrische Skala ist stetig.  
5. Fassungsvermögen von Binnenschiffen: Metrische Skala ist stetig.  
6. Ölverbrauch in einem Einfamilienhaus pro Jahr: Metrische Skala ist stetig.

## 3. Aufgabe

Ordne die gegebenen Merkmale zu A, B, D oder D
- A - Qualitatives Merkmal (Nominalskala)
- B - Qualitatives Merkmal (Ordinalskala)
- C - Quantitatives Merkmal (Kardinalskala, stetig)
- D - Quantitatives Merkmal (Kardinalskala, diskret)

**Merkmale**:<br>
Blutdruck, Seitenzahl eines Romans, Besucherzahl im Stadion, Konfession, Stimmenanteile bei einer Wahl, Klassenstärke, Anzahl der Geburten einer Frau, Körpergewicht,  Schulnoten, Schuhgröße, Krankenhausaufenthalt in Tagen, Hautfarbe, Staatsangehörigkeit.

### Lösung

A: Konfession, Haarfarbe, Staatsangehörigkeit  
B: Schulnoten, Schuhgröße
C: Blutdruck, Stimmenanteile, Körpergewicht
D: Seitenzahl, Besucherzahl, Klassenstärke, Geburtenzahl, Krankenhausaufenthalt

**Anmerkungen:**<br>
Schuhgröße wird oft als Beispiel für eine **ordinale Skala** verwendet, weil die Größen geordnet werden können (z.B. 36, 38, 40 usw.), aber die Abstände zwischen den Größen sind nicht unbedingt gleich groß. Zum Beispiel ist der Unterschied zwischen den Schuhgrößen 36 und 38 nicht zwingend der gleiche wie zwischen 38 und 40, obwohl sie numerisch aufeinanderfolgen.

## 4. Aufgabe

Sind folgende Aussagen richtig oder falsch:  
1. Die Merkmalsausprägungen von quantitativ-diskreten Merkmalen können auf einem bestimmten Intervall alle reellen Zahlen annehmen.  
2. Die Ordinalskala stellt die einfachste Form des Messens dar. Sie dient hauptsächlich zur Klassifizierung der Merkmalswerte.

### Lösung

1. Die Merkmalsausprägungen von quantitativ- diskreten Merkmalen können auf einem bestimmten Intervall alle reellen Zahlen annehmen.<br>
   Die Aussage ist falsch. Sie gilt nur bei quantitativ-stetigen Merkmalen. Bei diskreten Werten fehlen die Zwischenwerte.  

2. Die Ordinalskala stellt die einfachste Form des Messens dar. Sie dient hauptsächlich zur Klassifizierung der Merkmalswerte.<br>
   Die Aussage ist falsch. Die Nominalskala stellt die einfachste Form des Messens dar, nicht die Ordinalskala
