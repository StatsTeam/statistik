## 2. Aufgabe

Gib zu folgenden Merkmalen eine sinnvolle Anzahl von Merkmalsausprägungen an.  
1. Familienstand  
2. Nationalität  
3. Urlaubstage pro Jahr  
4. Entfernung zwischen Wohnung und Schule  
5. Automarke

### Lösung

1. Familienstand: ledig; verheiratet; verwitwet; geschieden  
2. Nationalität: deutsch; englisch; französisch; spanisch; türkisch; andere  
3. Urlaubstage pro Jahr: 0 bis unter 20; 20 bis unter 30; über 30  
4. Entfernung zwischen Wohnung und Schule: 0 – 50 km; oder in Klassen: 0 bis unter 2 km; 2 bis unter 10 km; über 10 km  
5. Automarke: Fiat; Mercedes; BMW

## 4. Aufgabe

Durch Befragung ist die Zufriedenheit der Klasse mit den unterrichtenden Lehrern zu untersuchen.  
- Welche Merkmale der Lehrer sind zur Beurteilung dieser Frage wichtig?  
- Lege sinnvolle Merkmalsausprägungen fest!

### Lösung

|      Merkmal      |                          Merkmalsausprägung                          |
| :---------------: | :------------------------------------------------------------------: |
|   Gerechtigkeit   |         Skala von 1 bis 3 (sehr gerecht; gerecht; ungerecht)         |
|   Fachkompetenz   |         Skala von 1 bis 6; 1 = hervorragend; 6 = ungenügend          |
|    Motivation     |          Skala von 1 bis 3 (gut, befriedigend, ungenügend)           |
|    Engagement     |               Skala von 1 bis 3 (groß; mittel; gering)               |
|     Toleranz      |       Skala von 1 bis 3 (sehr tolerant; tolerant; untolerant)        |
| Hilfsbereitschaft | Skala von 1 bis 3 (sehr hilfsbereit; hilfsbereit; nicht hilfsbereit) |
