```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Definieren der Klasse für die lineare Funktion
class LinearFunctionPlotter:
    def __init__(self, m, n):
        """
        Initialisiert die lineare Funktion mit Steigung m und y-Achsenabschnitt n.

        Parameters:
        - m (float): Die Steigung der Funktion.
        - n (float): Der y-Achsenabschnitt der Funktion.
        """
        self.m = m
        self.n = n

    def plot(self, points=None):
        """
        Erstellt einen Plot der linearen Funktion mit dem gegebenen Design
        und markiert optionale Punkte.

        Parameters:
        - points (list of tuples, optional): Eine Liste von Punkten (x, y),
          die auf dem Plot markiert werden sollen.
        """
        sns.set_palette("crest")  # Farbpalette für das Design
        
        # x-Werte und y-Werte berechnen
        x_values = np.linspace(-10, 10, 100)
        y_values = self.m * x_values + self.n

        # Erstellen des Plots
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.plot(x_values, y_values, label=f"f(x) = {self.m}x + {self.n}", color="blue")
        
        # Mittiges Koordinatensystem
        ax.spines['left'].set_position('zero')  # y-Achse bei x=0
        ax.spines['bottom'].set_position('zero')  # x-Achse bei y=0
        ax.spines['right'].set_color('none')  # Entfernen der rechten Achse
        ax.spines['top'].set_color('none')  # Entfernen der oberen Achse

        # Achsenbeschriftungen anpassen
        ax.yaxis.set_ticks_position('left')
        ax.xaxis.set_ticks_position('bottom')
        
        # Entferne den Ursprungspunkt 0 von den Achsenticks
        ax.set_xticks([tick for tick in ax.get_xticks() if tick != 0])
        ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])
        
        # Markiere die Punkte auf dem Plot
        if points:
            x_points, y_points = zip(*points)  # Entpacke die x- und y-Werte der Punkte
            ax.scatter(x_points, y_points, color="red", marker="o", label="Markierte Punkte")
        
        # Anpassungen für Stil und Beschriftungen
        ax.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
        ax.set_xlabel("x")
        ax.set_ylabel("f(x)")
        ax.set_title("Lineare Funktion f(x) = mx + n")
        ax.legend(loc="upper left")
        
        sns.despine()  # Entfernt die oberen und rechten Rahmenlinien für ein klareres Design
        plt.tight_layout()
        plt.show()

# Beispielinstanz der Klasse und Aufruf der Plot-Methode mit markierten Punkten
plotter = LinearFunctionPlotter(2, 3)
plotter.plot(points=[(-1, 1), (0, 3), (1, 5), (2, 7)])
```
