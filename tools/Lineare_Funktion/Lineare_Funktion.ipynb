```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Definieren der Klasse für die lineare Funktion
class LinearFunctionPlotter:
    def __init__(self, m, n):
        """
        Initialisiert die lineare Funktion mit Steigung m und y-Achsenabschnitt n.

        Parameters:
        - m (float): Die Steigung der linearen Funktion f(x) = mx + n.
        - n (float): Der y-Achsenabschnitt (Ordinatenabschnitt) der Funktion.
        """
        self.m = m
        self.n = n

    def plot(self, points=None, x_range=(-20, 20), second_function=None, x_min=None, x_max=None, x_step=None, y_min=None, y_max=None, y_step=None):
        """
        Zeichnet einen Plot der linearen Funktion f(x) = mx + n im angegebenen Design und 
        markiert optional übergebene Punkte. Eine zweite Funktion kann ebenfalls dargestellt werden.

        Parameters:
        - points (list of tuples, optional): Liste von Koordinatenpunkten [(x1, y1), (x2, y2), ...],
          die als markierte Punkte in der Grafik angezeigt werden sollen. Standard ist None.
        - x_range (tuple of floats, optional): Bereich der x-Werte als Tupel (x_min, x_max),
          der die Länge der gezeichneten Linie festlegt. Standard ist (-20, 20).
        - second_function (tuple, optional): Ein Tupel (m, n) zur Angabe der Steigung und des
          y-Achsenabschnitts der zweiten Funktion, die zusätzlich zur ersten Funktion geplottet wird.
          Standard ist None.
        - x_min (float, optional): Untere Grenze der x-Achse für die Tick-Beschriftung. 
          Nur relevant, wenn x_step gesetzt ist.
        - x_max (float, optional): Obere Grenze der x-Achse für die Tick-Beschriftung.
          Nur relevant, wenn x_step gesetzt ist.
        - x_step (float, optional): Schrittweite der Tick-Beschriftungen entlang der x-Achse.
          Wird nur verwendet, wenn x_min und x_max definiert sind.
        - y_min (float, optional): Untere Grenze der y-Achse für die Tick-Beschriftung.
          Nur relevant, wenn y_step gesetzt ist.
        - y_max (float, optional): Obere Grenze der y-Achse für die Tick-Beschriftung.
          Nur relevant, wenn y_step gesetzt ist.
        - y_step (float, optional): Schrittweite der Tick-Beschriftungen entlang der y-Achse.
          Wird nur verwendet, wenn y_min und y_max definiert sind.

        Notes:
        - Das Plot-Design umfasst eine mittige Platzierung der Achsen, sodass die Achsen
          durch den Ursprung (0,0) verlaufen.
        - Pfeile an den Enden der x- und y-Achse werden hinzugefügt, um die Richtung
          eines kartesischen Koordinatensystems darzustellen.
        - Eine optionale zweite Funktion kann zum Vergleich hinzugefügt werden, falls
          der Parameter `second_function` gesetzt ist.
        - Zusätzliche Parameter für die x- und y-Ticks erlauben eine flexible Anpassung
          der Achsenbeschriftungen.
        """
        sns.set_palette("crest")  # Farbpalette für das Design
        
        # x-Werte und y-Werte für die erste Funktion berechnen
        x_values = np.linspace(x_range[0], x_range[1], 100)
        y_values = self.m * x_values + self.n

        # Erstellen des Plots
        fig, ax = plt.subplots(figsize=(10, 6))
        ax.plot(x_values, y_values, label=f"f(x) = {self.m}x + {self.n}", color="blue")

        # Zweite Funktion plotten, falls angegeben
        if second_function is not None:
            m2, n2 = second_function
            y_values_2 = m2 * x_values + n2
            ax.plot(x_values, y_values_2, label=f"g(x) = {m2}x + {n2}", color="green")

        # Mittiges Koordinatensystem
        ax.spines['left'].set_position('zero')  # y-Achse bei x=0
        ax.spines['bottom'].set_position('zero')  # x-Achse bei y=0
        ax.spines['right'].set_color('none')  # Entfernen der rechten Achse
        ax.spines['top'].set_color('none')  # Entfernen der oberen Achse

        # Achsenbeschriftungen anpassen
        ax.yaxis.set_ticks_position('left')
        ax.xaxis.set_ticks_position('bottom')
        
        # Pfeile für x- und y-Achse hinzufügen
        ax.plot((1), (0), ls="", marker=">", ms=10, color="black", transform=ax.get_yaxis_transform(), clip_on=False)
        ax.plot((0), (1), ls="", marker="^", ms=10, color="black", transform=ax.get_xaxis_transform(), clip_on=False)
        
        # Entferne den Ursprungspunkt 0 von den Achsenticks
        ax.set_yticks(np.arange(y_min, y_max, y_step))
        ax.set_xticks(np.arange(x_min, x_max, x_step))
        ax.set_xticks([tick for tick in ax.get_xticks() if tick != 0])
        ax.set_yticks([tick for tick in ax.get_yticks() if tick != 0])

        # Markiere die Punkte auf dem Plot
        if points:
            x_points, y_points = zip(*points)  # Entpacke die x- und y-Werte der Punkte
            ax.scatter(x_points, y_points, color="red", marker="o", label="Markierte Punkte")
        
        # Anpassungen für Stil und Beschriftungen
        ax.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
        ax.text(1.05, 0, "x", transform=ax.get_yaxis_transform(), ha="left", va="center", fontsize=12)
        ax.text(0, 1.05, "f(x)", transform=ax.get_xaxis_transform(), ha="center", va="bottom", fontsize=12)
        ax.set_title(r"$\bf{Lineare\ Funktion:}$ $f(x) = mx + n$", fontsize=16, color="darkblue", pad=20, loc="center")
        ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
        plt.tight_layout()

```

```python
plotter = LinearFunctionPlotter(11, -4500)
plotter.plot( x_range=(0, 800), second_function=(-5, 1000), x_min=0, x_max=800, x_step=50, y_min=-5000, y_max=5000, y_step=1000)
```
