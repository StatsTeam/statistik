```python
import matplotlib.pyplot as plt
import numpy as np


def create_multiple_boxplots(data_lists, labels=None, title="Title", y_label="y-values"):
    """
    Erstellt mehrere Boxplots nebeneinander mit dem ursprünglichen Designstil und zeigt die wichtigsten Werte an.
    
    Parameter:
    - data_lists: Liste von Listen mit Datenpunkten für die einzelnen Boxplots
    - labels: Liste von Labels für jeden Boxplot (optional)
    - title: Titel des Diagramms (optional)
    - y_label: Bezeichnung der y-Achse (optional)
    """
    fig, ax = plt.subplots(figsize=(12, 8))
    fig.patch.set_facecolor('#FFFBE9')
    
    # Boxplots für die Datenliste erstellen
    bplot = ax.boxplot(
        data_lists,
        patch_artist=True,
        medianprops=dict(color='#F07B3F', linewidth=2),
        flierprops=dict(marker='o', markerfacecolor='#FF2E63', markersize=8, linestyle='none')
    )
    
    # Farben für jeden Boxplot festlegen
    colors = ['#3282B8', 'orange', 'tomato', 'green', 'purple']
    for patch, color in zip(bplot['boxes'], colors):
        patch.set_facecolor(color)
    
    # Optional: Labels für die Boxplots festlegen
    if labels:
        ax.set_xticks(range(1, len(data_lists) + 1))
        ax.set_xticklabels(labels)
    
    ax.set_title(title, fontweight='bold')
    ax.set_ylabel(y_label)
    ax.set_facecolor('#FFFBE9')

    # Berechnung und Anzeige von Quartilen, Median, Min und Max für jeden Boxplot
    for i, data in enumerate(data_lists):
        q1_value = np.percentile(data, 25)
        q3_value = np.percentile(data, 75)
        median_value = np.median(data)
        
        # Min und Max ohne Ausreißer-Grenzen
        min_value = np.min(data)
        max_value = np.max(data)

        # Verschiebung der Texte nach rechts, z.B. von 1.15 auf 1.25
        shift_x = 1.25
        ax.text(i + shift_x, q1_value, f'Q1: {q1_value:.2f}', verticalalignment='center', fontsize=10, color='#3282B8', fontweight='bold')
        ax.text(i + shift_x, median_value, f'Median: {median_value:.2f}', verticalalignment='center', fontsize=10, color='#F07B3F', fontweight='bold')
        ax.text(i + shift_x, q3_value, f'Q3: {q3_value:.2f}', verticalalignment='center', fontsize=10, color='#3282B8', fontweight='bold')
        ax.text(i + shift_x, min_value, f'Min: {min_value:.2f}', verticalalignment='center', fontsize=10, color='#3282B8', fontweight='bold')
        ax.text(i + shift_x, max_value, f'Max: {max_value:.2f}', verticalalignment='center', fontsize=10, color='#3282B8', fontweight='bold')
    
    # Rahmenlinien und Gitter entfernen
    ax.spines['top'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.yaxis.set_ticks_position('left')
    ax.grid(True, linestyle="--", alpha=0.7, axis="y")
```

```python
data1 = [6, 6, 7, 8, 11, 13, 15, 15, 14, 10, 8, 7]
data2 = [9, 1, 4, 8, 13, 16, 20, 18, 14, 8, 3, 0]
data3 = [-10, -9, -4, 5, 12, 17, 20, 18, 11, 4, -2, -8]
data4 = [9, 11, 13, 15, 19, 23, 27, 27, 24, 19, 13, 10]
data_lists = [data1, data2, data3, data4]
labels = ["Region A", "Region B", "Region C", "Region D"]

create_multiple_boxplots(data_lists, labels=labels, title="Temperaturen", y_label="Temperatur in Grad Celsius")
```
