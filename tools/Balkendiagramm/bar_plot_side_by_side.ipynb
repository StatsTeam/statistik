```python
import matplotlib.pyplot as plt
import numpy as np

# Daten für Aufgabe 1 und Aufgabe 2
monate = ["Jan", "Feb", "Mär", "Apr", "Mai", "Jun", "Jul", "Aug", "Sep", "Okt", "Nov", "Dez"]
temperaturen_aufgabe1 = [4.3, 5.7, 5.8, 9.6, 14.0, 18.1, 20.5, 19.8, 17.7, 9.5, 5.7, 0.3]
temperaturen_aufgabe2 = [0.5, 2.9, 3.7, 7.5, 13.3, 17.1, 18.7, 17.8, 17.1, 8.2, 3.6, 1.5]

# Berechnung der arithmetischen Mittelwerte
mittelwert_aufgabe1 = np.mean(temperaturen_aufgabe1)
mittelwert_aufgabe2 = np.mean(temperaturen_aufgabe2)

# Balkenbreite und Positionen für die Gruppen
bar_width = 0.35
index = np.arange(len(monate))

# Erstellung des Balkendiagramms
fig, ax = plt.subplots(figsize=(10, 6))

balken1 = ax.bar(
    index - bar_width / 2,
    temperaturen_aufgabe1,
    bar_width,
    label='Aufgabe 1',
    color="dodgerblue"
    )

balken2 = ax.bar(
    index + bar_width / 2,
    temperaturen_aufgabe2,
    bar_width,
    label='Aufgabe 2',
    color="orange"
    )

# Einfügen der Mittelwerte als horizontale Linien
ax.axhline(y=mittelwert_aufgabe1, color='dodgerblue', linestyle='--', linewidth=1, label=f'Arithm. Mittel Aufgabe 1: {mittelwert_aufgabe1:.2f}°C')
ax.axhline(y=mittelwert_aufgabe2, color='orange', linestyle='--', linewidth=1, label=f'Arithm. Mittel Aufgabe 2: {mittelwert_aufgabe2:.2f}°C')

# Diagrammanpassungen
ax.set_xlabel('Monat')
ax.set_ylabel('Temperatur (°C)')
ax.set_title('Monatliche Temperaturen für Aufgabe 1 und Aufgabe 2')
ax.set_xticks(index)
ax.set_xticklabels(monate)
ax.grid(color="gray", linestyle="--", linewidth=0.5, alpha=0.7)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))
plt.tight_layout()
```
