# Venn-Diagram

```python
import matplotlib.pyplot as plt
import random
from adjustText import adjust_text

def draw_venn_diagram(elements, title="Venn-Diagramm der Menge", formula=None):
    # Erstellen der figure und axes Objekte
    fig, ax = plt.subplots(figsize=(6, 6), facecolor='white')

    # Zeichnen des Kreises mit einem minimalistischen Design
    circle = plt.Circle((0.5, 0.5), 0.4, color='dodgerblue', edgecolor='#5dade2', linewidth=1.5)
    ax.add_patch(circle)

    # Achseneinstellungen für ein sauberes, zentriertes Layout
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')  # Entferne die Achsen für ein minimalistisches Design

    # Liste zur Speicherung der Textobjekte für die automatische Anpassung
    texts = []

    # Zufällige Positionierung der Elemente im Kreis ohne Überlappung
    for element in elements:
        while True:
            # Generiere zufällige x, y Koordinaten innerhalb des Kreises
            x, y = random.uniform(0.3, 0.7), random.uniform(0.3, 0.7)
            # Überprüfen, ob die Position innerhalb des Kreises liegt
            if (x - 0.5) ** 2 + (y - 0.5) ** 2 <= 0.4 ** 2:
                # Füge den Text zur Liste hinzu, damit adjust_text ihn anpassen kann
                text = ax.text(x, y, str(element), fontsize=13, ha='center', va='center', color="white")
                texts.append(text)
                break

    # Platzieren Sie die Formel in der Mitte des Kreises, falls eine Formel übergeben wurde
    if formula:
        ax.text(0.5, 0.5, f"${formula}$", fontsize=20, ha='center', va='center', color="white")

    # Passen Sie die Position der Texte an, um Überschneidungen zu vermeiden
    adjust_text(texts, only_move={'points': 'xy', 'texts': 'xy'}, autoalign='xy', force_text=0.5, expand_text=(1.2, 1.2))

    # Titel setzen und im minimalistischen Stil anzeigen
    ax.set_title(title, fontsize=16, color='#34495e', pad=20, fontweight="bold")
```

## 1. Beispiel

```python
elements = {}
formula = r"U=\{1, 3, 5, 7, 9, ...\}"
draw_venn_diagram(elements, title="Venn-Diagramm der Menge U", formula=formula)
```
