```python
import numpy as np
import matplotlib.pyplot as plt

def plot_histogram(data, bins=20, highlight_value=0.6, xlabel="Proportion of Male Births", ylabel="Relative Frequency", title=None):
    """
    Plots a histogram with relative frequencies and highlights the area starting from a specific value.

    Parameters:
    - data: Array-like, data to plot.
    - bins: int, number of bins for the histogram (default=20).
    - highlight_value: float, value on the x-axis from where to highlight the area (default=0.6).
    - xlabel: str, label for the x-axis.
    - ylabel: str, label for the y-axis.
    - title: str, title of the histogram.
    """
    counts, bin_edges = np.histogram(data, bins=bins, density=True)

    # Plot histogram
    plt.bar(bin_edges[:-1], counts, width=np.diff(bin_edges), align='edge', color="blue", alpha=0.7, edgecolor="black", label="Data")

    # Highlight area starting from highlight_value
    plt.fill_betweenx(y=[0, max(counts)], x1=highlight_value, x2=max(bin_edges), color='red', alpha=0.3, label=f'Highlight >= {highlight_value}')
    plt.axvline(x=highlight_value, color='red', linestyle='--', label=f'Highlight: {highlight_value}')

    # Set X-axis ticks to bin edges
    plt.xticks(bin_edges, rotation=45)

    # Add labels and title
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title if title else "Histogram")
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage:
# Simulate data for small hospital
small_hospital_data = np.random.binomial(15, 0.5, 10000) / 15
plot_histogram(small_hospital_data, bins=15, highlight_value=0.6, xlabel="Proportion of Male Births", ylabel="Relative Frequency", title="Small Hospital: Male Birth Proportions")

large_hospital_data = np.random.binomial(45, 0.5, 10000) / 45
plot_histogram(large_hospital_data, bins=15, highlight_value=0.6, xlabel="Proportion of Male Births", ylabel="Relative Frequency", title="Large Hospital: Male Birth Proportions")
```
