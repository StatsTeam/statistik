# Das gewichtete arithmetische Mittel

Mit einem Einführungsbeispiel möchte ich das gewichtete arithmetische Mittel vorstellen:

In einer Badewanne befinden sich $v_1=150\text{ Liter}$ Wasser der Temperatur  $t_1=50\degree C$. Nun schüttet man $v_2=5\text{ Liter}$ Wasser mit der Temperatur $t_2=20\degree C$ dazu. Welche Mischungstemperatur $t_m$ stellt sich ein?
<br>
<br>
Es ist klar, dass die Mischungstemperatur nicht das arithmetische Mittel der Temperaturen $t_1$ und $t_2$ ist, denn die Wassermengen sind unterschiedlich. Somit werden die einzelnen Temperaturen unterschiedlich stark gewichtet.
<br>
<br>
Aus der Physik ist die folgende Berechnung der Mischungstemperatur bekannt:

$$t_m=\frac{m_1t_1+m_2t_2}{m_1+m_2}=\frac{150\cdot50+5\cdot20}{150+5}\approx49$$

Dabei ist der Term für $t_m$ das sogenannte gewichtete arithmetische Mittel!

## Definition

Das gewichtete arithmetische Mittel ist eine Methode zur Berechnung des Durchschnitts, bei der jedem Datenpunkt ein bestimmtes "Gewicht" zugewiesen wird. 
<br>
<br>
Im Gegensatz zum einfachen arithmetischen Mittel, bei dem alle Datenpunkte gleich gewichtet werden, werden beim gewichteten arithmetischen Mittel einige Datenpunkte stärker gewichtet als andere, basierend auf ihrem Einfluss auf den Gesamtdurchschnitt.

> Definition: $$\overline{x_{gew}}=\frac{\sum_{i=1}^ng_i\cdot{x_i}}{\sum_{i=1}^ng_i} $$
> mit den Gewichten $g_1...g_n$ wobei gilt $g_i \geq 0$ 

Das gewichtete arithmetische Mittel wird in vielen Situationen verwendet, in denen Werte unterschiedlich wichtig oder unterschiedlich relevant sind. Beispiele hierfür sind:
- Notensysteme in der Bildung: Wenn Prüfungen, Hausaufgaben und Projekte unterschiedliche Gewichtungen haben, fließt jede Leistung entsprechend ihrer Relevanz in die Endnote ein.
- Finanzen und Aktienportfolios: Beim Berechnen des durchschnittlichen Kaufpreises von Aktien, wenn Aktien zu unterschiedlichen Preisen und in unterschiedlichen Mengen gekauft wurden, gewichtet man den Preis mit der Anzahl der gekauften Aktien.
- Statistische Analysen und Durchschnittspreise: Zum Beispiel bei einem Warenkorbpreisindex, bei dem verschiedene Produkte unterschiedlich häufig gekauft werden, sodass häufiger gekaufte Artikel mehr Einfluss auf den Durchschnittspreis haben.
- Mengenverhältnisse und Mischungen: In der Chemie und Physik, wenn Mischungen berechnet werden, werden Komponenten entsprechend ihrer Anteile im Gesamtvolumen gewichtet.

Das gewichtete Mittel verwendet im Kern dieselbe Logik wie das einfache arithmetische Mittel – nur dass hier die Werte mit unterschiedlichen Gewichten multipliziert werden. Das Ergebnis ist ein anderes arithmetisches Mittel, da die Berechnung die Gewichte berücksichtigt.

## Anwendungsbeispiel

Ein Autofahrer, der beruflich viel unterwegs ist, hat im letzten Monat dreimal getankt. Beim ersten Mal waren es $50$ Liter für $1,14$€ pro Liter, dann $30$ Liter für $1,21$€ pro Liter und schließlich 55 Liter für 1,19€ pro Liter.
Welchen durchschnittlichen Preis pro Liter hat der Autofahrer bezahlt?
<br>
<br>
In diesem Beispiel ist das Tankvolumen bzw. die Liter, die gewichte. Sie sagen also aus, welcher Preis wie stark gewichtet wird:

$g_1=50\text{, }g_2=30\text{, }g_3=55$

Der Gesamtpreis beträgt: 

$50\cdot1,14+30\cdot1,21+55\cdot1,19=158,75\text{ Euro}$

Insgesamt wurde getankt: 

$50+30+55=135\text{ Liter}$

Nun kann man den durchschnittlichen Liter-Preis berechnen:

$$\overline{x_{gew}}=\frac{50\cdot1,14+30\cdot1,21+55\cdot1,19}{50+30+55}\approx1,176\text{ €/Liter}$$

## Abschließende Bemerkung

Der Name "gewichtetes arithmetisches Mittel" darf nicht den Eindruck erwecken, dass es sich dabei um einen anderen Mittelwert als das arithmetische Mittel handelt. Es liegt nur eine andere Art der Berechnung vor.
<br>
<br>
Beim gewichteten Mittel liegen die Daten als Häufigkeitsverteilung vor, beim arithmetischen Mittel liegen dagegen Einzelbeobachtungen vor.

**Anwendungsbeispiel:**

Ein Unternehmen hat folgende Einkäufe zu verbuchen:
- 01.02. – $20$ Sack Zement zum Preis von $12,00$ € / Sack
- 03.04. – $35$ Sack Zement zum Preis von $15,00$ €/ Sack
- 07.05. – $123$ Sack Zement zum Preis von $9,00$ €/ Sack

Wie ist der durchschnittliche Preis pro Sack, wenn die Menge mit berücksichtigt wird?

1. Berechnung durch das gewichtete arithmetische Mittel:

    $\overline{x_{gew}}=\frac{\sum_{i=1}^ng_i\cdot{x_i}}{\sum_{i=1}^ng_i} $

    Dabei sind die Gewichtungen, die Anzahl der Säcke:

    $\overline{x_{gew}}=\frac{20\cdot 12 + 35\cdot 15 + 123\cdot 9}{20+35+123}\approx 10.52 €\text{ pro Sack}$

2. Berechnung durch das arithmetische Mittel:
   
    $\overline{x}=\frac{1}{n}\cdot\sum_{i=1}^nx_i$

    In unserem Fall sind keine Einzelwerte, sondern eine Häufigkeitsverteilung vorhanden:

    $x_i=n_i\cdot x_i$

    Somit wird das arithmetische Mittel bei einer Häufigkeitsverteilung berechnet durch:

    $\overline{x}=\frac{1}{n}\cdot\sum_{i=1}^n(n_i\cdot x_i)$

    Im Endeffekt ist dies nichts weiter als das arithmetische gewichtete Mittel:

    $\sum_{i=1}^n(n_i\cdot x_i)=20\cdot 12 + 35\cdot 15 + 123\cdot 9=1872€$

    $n=20+35+123=178 \text{ Säcke}$

    $\overline{x}=\frac{1872€}{178 \text{ Säcke}}=\approx 10.52 €\text{ pro Sack}$

Allgemein formuliert lässt sich das arithmetische Mittel für Häufigkeitsverteilungen berechnen:

| Reihe Nr.             | Anzahl der Messwerte | Arithmetisches Mittel |
| --------------------- | -------------------- | --------------------- |
| 1                     | $n_1$                | $\overline{x_1}$      |
| 2                     | $n_2$                | $\overline{x_2}$      |
| ...                   | ...                  | ...                   |
| k                     | $n_k$                | $\overline{x_k}$      |
| Zusammengesetze Reihe | $n=n_1+n_2+...+n_k$  | $\overline{x}$        |

<br>    
    
> Definition des arithmetischen mittels für Häufigkeitsverteilung: $$\overline{x}=\frac{n_1\cdot\overline{x_1}+n_2\cdot\overline{x_2}+...+n_k\cdot\overline{x_k}}{n_1+n_2+...+n_k}=\frac{1}{n}\cdot\sum_{i=1}^nn_i\cdot x_i$$
