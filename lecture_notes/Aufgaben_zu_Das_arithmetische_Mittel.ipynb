# Aufgaben zu: Das arithmetische Mittel

## 1. Aufgabe

Durch die Bildung des arithmetischen Mittels lässt sich der jährliche Mittelwert der Temperatur bestimmen. Jedoch vernachlässigt man die Tatsache dass die Monate unterschiedliche Längen haben. Berechnen Sie das arithmetische Mittel der Temperatur!

| Monat      | Jan  | Feb  | Mär  | Apr | Mai | Jun  |
| ---------- | ---- | ---- | ---- | --- | --- | ---- |
| Temperatur | 4,3  | 5,7  | 5,8  | 9,6 | 14  | 18,1 |
| Monat      | Jul  | Aug  | Sep  | Okt | Nov | Dez  |
| Temperatur | 20,5 | 19,8 | 17,7 | 9,5 | 5,7 | 0,3  |

### Lösung

$\overline{x}=\frac{(4,3+5,7+...5,7+0,3)}{12}=10,9$

## 2. Aufgabe

Bestimmen Sie den Jahresmittelwert und stellen Sie die Verteilungen aus Aufgabe 1 und Aufgabe 2 in ein gemeinsames Diagramm graphisch dar.

## 3. Aufgabe

Zeigen Sie die Gültigkeit der Aussage (Schwerpunkteigenschaft):

$$(x_1-\overline{x})+(x_2-\overline{x})+...+(x_n-\overline{x})=0$$

### Lösung

Das arithmetische Mittel lautet:

$$\overline{x}=\frac{x_1+...+x_n}{n}$$

Nun mupltiplizier man diese Gleichung mit $n$, es folgt:

$$n\cdot\overline{x}=x_1+...+x_n$$

Jetzt kann man den linken Term der Gleichung umschreiben, indem man $n\cdot\overline{x}$ durch Addition ersetzt:

$$\overline{x}+...+\overline{x}=x_1+...+x_n$$

Jetzt muss man nur noch von jeder Zahl das arithmetische Mittel subtrahieren:

$$x_1-\overline{x}+...+x_n-\overline{x}=0$$

Wenn wir die Klammern setzen, sieht es genau so aus wie die Schwerpunkteigenschaft:

$$(x_1-\overline{x})+...+(x_n-\overline{x})=0$$

## 4. Aufgabe

Ein Kochkurs für Männer besteht aus 5 Männern, die Kochen lernen wollen und folgende Alter haben : `18, 20, 20, 22, 25`.
Berechnen sie das arithmetische Mittel, mit den 5 Teilnehmern!
<br>
<br>
Nun tritt dem Kochkurs ein 75-jähriger bei, berechnen sie jetzt das arithmetische Mittel mit dem neuen Teilnehmer inklusive.
<br>
<br>
Vergleichen Sie die beiden arithmetischen Mittelwärte. Was fällt Ihnen auf, im Bezug auf Ausreißer? Ist das arithmetische Mittel immer noch gut wenn der 75-jährige Teilnehmer sich dem Kurs anschließt?

### Lösung

- 5 Teilnehmer:
  $\overline{x}=\frac{18+20+20+22+25}{5}=21$
- 6 Teilnehmer:
  $\overline{x}=\frac{18+20+20+22+25+75}{6}=30$
- Beobachtung:
  Der Wert von $75$ stellt einen Ausreißer dar, da er deutlich höher ist als die Altersgruppe der übrigen Teilnehmer. Dieser hohe Wert hat das arithmetische Mittel stark nach oben verschoben. Das arithmetische Mittel ist empfindlich gegenüber Ausreißern und kann in diesem Fall ein falsches Bild des "Durchschnittsalters" im Kurs vermitteln. Es ist weniger geeignet, die Altersstruktur der Gruppe darzustellen, da das Alter von $30$ Jahren die Realität der jüngeren Gruppe nicht gut repräsentiert.

## 5. Aufgabe

Folgende Daten sind gegeben: `2, 5, 12, 20`. Nun soll ein fünfter Datenpunkt gefunden werden, so dass der arithmetische Mittelwert $\overline{x}=9$ wird.

### Lösung

$\overline{x}=\frac{2+5+12+20+x_5}{5}$

Umstellen nach $x_5$ liefert den fünften gesuchten Wert:

$5\cdot\overline{x}=2+5+12+20+x_5=39+x_5$

$x_5=5\cdot\overline{x}-39=5\cdot\overline{9}-39=6$
