Eine Maschinenfabrik produziert die Ketten für Kettensägen. Das Einrichten der einzelnen notwendigen Maschinen kostet $4500$ €, die Herstellung jeder Kette $9$ €.
<br>
Du erkennst, dass die Kosten der Ketten abhängig von der Anzahl der Ketten sind.
<br>
Diese Kosten sind variabel: Je mehr Ketten, desto höher die Kosten.
<br>
Der Einrichtungspreis der Maschinen ist fix. Er ändert sich nicht.
<br>
<br>
So heißt die Funktion für diesen Sachverhalt: 
- $k(x)=9x+4500$
- Mit $x$ als Anzahl der Ketten
- Mit $k$ als Kosten

>**Füge hier nun die Grafik der Funktion $k(x)$ hinzu:**

Für den Fabrikchef ist aber vor allem der Gewinn interessant.
Dazu berechnet er erstmal den Umsatz. Das ist das Geld, das er durch den Verkauf der Ketten einnimmt.
Nach zahlreichen Recherchen setzt der Chef den Verkaufspreis von $20$ € pro Kette an.
<br>
<br>
Aus dem Sachverhalt ergibt sich folgende Funktion: 
- $u(x)=20x$
- Mit $x$ für Anzahl der Ketten
- Mit $u$ für Umsatz in Euro

>**Füge hier nun die Grafik der Funktion $u(x)$ hinzu:**

Frage: Wie viele Ketten müssen hergestellt werden, damit die Firma einen Gewinn erzielt?
>**Zeichne beide Graphen $k(x)$ und $u(x)$ in ein Koordinatensystem, und füge sie an der folgenden Stelle ein:**

Aus der Grafik erkennt man, dass der Umsatz bei ungefähr 400 Stück, über den Kosten liegt.
Das heißt: Ab dieser Stückzahl erzielt das Unternehmen einen Gewinn.
<br>
<br>
Dabei handelt es sich um den Schnittpunkt der beiden Geraden, um ihn zu berechnen muss man die beiden Funktionsgleichungen gleichsetzen:
$$u(x)=k(x)$$

Aus der Lösung dieser Gleichung folgt:

$9x+4500=20x$

$x=409.1$

Der errechnete Wert bedeutet, dass ab 410 verkauften Ketten der Umsatz größer ist als die Kosten: die Firma macht einen Gewinn.
<br>
<br>
Für den Gewinn ist auch eine Funktionsgleichung praktisch: Ziehe vom Umsatz die Kosten ab.

$g(x)=u(x)-k(x)$

$g(x)=20x-(9x+4500)$

$g(x)=11x-4500$

>**Füge hier die Grafik der Funktion $g(x)$ hinzu:**

Du siehst, dass die Gerade bei etwas über 400 Stück die x-Achse schneidet.
Bei einer geringeren Stückzahl ist der Gewinn negativ (Verlust), danach positiv (Gewinn).
Die Vermutung liegt nahe, dass der Schnittpunkt bei $x=409$, liegt. Und das ist exakt der Schnittpunkt von $u(x)$ und $k(x)$.
