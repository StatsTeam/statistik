# Die Grundlagen der Arithmetik

Die Arithmetik ist der grundlegende Zweig der Mathematik, der sich mit Zahlen und grundlegenden Rechenoperationen befasst. Sie ist die Grundlage für fast alle weiteren Bereiche der Mathematik und bildet die Basis für alltägliche Berechnungen. Die wichtigsten Konzepte und Rechenoperationen der Arithmetik sind Addition, Subtraktion, Multiplikation und Division.

## Grundoperationen der Arithmetik

Hier werden die vier grundlegenden Rechenoperationen, die in der Arithmetik verwendet werden, im Detail betrachtet:

### Addition

Die Addition beschreibt das Zusammenfügen von zwei oder mehr Zahlen. Dabei sind die Zahlen welche miteinander addiert werden sogenannte Summanden und das Ergebnis ist die Summe:
- Schreibweise: $a+b=c$
  - $a$ und $b$ sind Summanden
  - $c$ ist die Summe

Bei der Addition gilt das sogenannte Kommutativgesetz und Assoziativgesetz:

> Kommutativgesetz: Die Reihenfolge der Summanden spielt bei der Addition keine Rolle $a+b=b+a$

> Assoziativgesetz: Bei der Addition mehrerer Zahlen kann die Reihenfolge der Berechnung geändert werden, ohne dass das Ergebnis beeinflusst wird $(a+b)+c=a+(b+c)$. Die Klammern geben an, welche Addition Vorrang hat. 

### Subtraktion 

Die Subtraktion beschreibt das Entfernen einer Zahl von einer anderen. Die Zahl, von der abgezogen wird, nennt man den Minuenden, die abzuziehende Zahl den Subtrahenden, und das Ergebnis heißt Differenz: 
- Schreibweise: $a-b=c$
  - $a$ ist der Minuend
  - $b$ ist Subtrahend
  - $c$ ist die Differenz

Im Gegensatz zur Addition ist die Subtraktion Nicht-Kommutativ und Nicht-Assoziativ:

> Die Subtraktion ist nicht kommutativ, was bedeutet, dass die Reihenfolge der Zahlen das Ergebnis beeinflusst $a-b \neq b-a$

> Die Subtraktion ist auch nicht assoziativ, was bedeutet, dass die Reihenfolge der Berechnung bei mehreren Subtraktionen das Ergebnis verändert $(a-b)-c \neq a-(b-c)$

### Multiplikation

Die Multiplikation beschreibt das Wiederholte Addieren derselben Zahl. Die Zahlen, die multipliziert werden, nennt man Faktoren, und das Ergebnis heißt Produkt:
- Schreibweise: $a\cdot b = c$
  - $a$ und $b$ sind Faktoren
  -  $c$ ist das Produkt

Bei der Multiplikation gilt, wie bei der Addition, das Kommutativgesetz und Assoziativgesetz. Es kommt jedoch ein neues Gesetz hinzu, das sogenannte Distributivgesetz:

> Die Reihenfolge der Faktoren spielt bei der Multiplikation keine Rolle $a \cdot b=b \cdot a$

> Bei der Multiplikation mehrerer Zahlen kann die Reihenfolge der Berechnung geändert werden, ohne dass das Ergebnis beeinflusst wird $(a \cdot b) \cdot c = a \cdot (b \cdot c)$

> Distributivgesetz: Die Multiplikation verteilt sich über die Addition und Subtraktion $a \cdot (b + c) = a \cdot b + a \cdot c$

> Außerdem ist die Multiplikation mit 0 ebenfalls definiert als $a \cdot 0 = 0$. Die Multiplikation mit Null, bedeutet dass man $a$ nullmal addiert. Wenn man etwas nullmal addiert, gibt es nichts zu addieren, daher ergibt sich das Produkt $0$.

### Division 

Die Division beschreibt das Aufteilen einer Zahl in gleiche Teile. Dabei ist die Zahl welche geteilt wird der Dividend, die Zahl durch welche geteilt wird der Divisor und das Resultat der Quotient:
- Schreibweise: $\frac{a}{b}=c$
  - $a$ ist der Dividend
  - $b$ ist Divisor
  - $c$ ist die Quotient

Wie sie Subtraktion ist diese Rechenoperation Nicht-Kommutativ und Nicht-Assoziativ:

> Nicht-Kommutativ: $\frac{a}{b}=\frac{b}{a}$

> Nicht-Assoziativ: $a \div (b \div c) \neq (a \div b) \div c$

> Die Division durch Null ist nicht definiert, weil es keinen sinnvollen Wert für eine Zahl $\frac{a}{0}$ gibt. Diese Operation führt zu keinem bestimmten Ergebnis, da das Konzept der „Aufteilung in null Teile“ mathematisch keinen Sinn ergibt. Bei der Division fragt man im Wesentlichen: „Wie oft passt der Divisor in den Dividend?“. Bei einer Division durch Null , würde man also fragen: „Wie oft passt Null in $a$?“. Diese Frage ergibt keinen Sinn, weil jede Zahl mit Null multipliziert immer Null ergibt und nie den Dividend $a$.

## Potenzen

Potenzen sind ein grundlegendes Konzept in der Mathematik, das verwendet wird, um wiederholte Multiplikationen kompakt zu notieren. Eine Potenz besteht aus einer Basis und einem Exponent und beschreibt, wie oft die Basis mit sich selbst multipliziert wird.

**Aufbau einer Potenz:**

Die Potenz wird in der Form $a^n$ geschrieben, wobei:
- $a$ die Basis ist (die Zahl welche potenziert wird) und
- $n$ der Exponent, die Anzahl der Multiplikationen mit sich selbst.

Beispielweise bedeutet $3^4=3 \cdot 3 \cdot 3 \cdot 3 = 81$ oder $2^8=256$ ($2$ wird $8$ mal mit sich selbst multipliziert).

**Besondere Fälle:**

- Exponent $1$:
  - Jede Zahl mit dem Exponenten 1 bleibt unverändert, weil sie nur einmal multipliziert wird.
  - Beispiel: $5^1=5$
- Exponent $0$:
  - Jede Zahl (außer 0), die auf die Potenz 0 erhoben wird, ergibt 1.
  - Beispiel: $7^0=1$
- Exponent $2$:
  - Wenn eine Zahl mit sich selbst multipliziert wird, spricht man von der Quadratzahl oder dem Quadrat.
  - Beispiel: $4^2=16$
- Exponent $3$:
  - Wenn eine Zahl dreimal mit sich selbst multipliziert wird, nennt man das die Kubikzahl oder den Kubus.
  - Beispiel: $2^3=8$

**Wissenschaftliche Schreibweise:**

Es handelt sich um eine Methode, um sehr große oder sehr kleine Zahlen kompakt und übersichtlich darzustellen. Diese Schreibweise wird häufig in der Wissenschaft und Mathematik verwendet, insbesondere bei Berechnungen mit extremen Zahlen, um Lesbarkeit und Genauigkeit zu verbessern.
<br>
<br>
In der wissenschaftlichen Schreibweise wird eine Zahl als Produkt einer Dezimalzahl (mit einer Ziffer vor dem Dezimalpunkt) und einer Potenz von 10 geschrieben:

$$a \cdot 10^n$$

Der Exponent kann sowohl positiv als auch negativ sein:
- Ein positiver Exponent bedeutet dass das Komma nach rechts verschoben wird, um $n$ Stellen
- Ein negativer Exponent bedeutet dass das Komma nach links verschoben wird, um $n$ Stellen

Beispiele sind:
- $3.5 \cdot 10^4 = 35000$
- $2.5 \cdot 10^{-3} = 0.0025$
- Kilo steht für $1000$ oder $10^3$
- Mega steht für $1000000$ oder $10^6$
- Mikro steht für $0.000001$ oder $10^{-6}$

## Brüche und Dezimalzahlen

Ein Bruch ist eine Darstellung für Teile eines Ganzen. Jeder Bruch besteht aus zwei Zahlen:
- Zählen (die Zahl auf dem Bruchstrich)
- Nenner (die Zahl unter dem Bruchstrich)

Geschrieben wird ein Bruch als:

$$\frac{a}{b}$$ 

Wobei $a$ der Zählen und $b$ der Nenner ist. Der Zähler gibt die Anzahl der Teile an, die betrachtet werden, während der Nenner die Gesamtzahl der Teile darstellt, in die das Ganze aufgeteilt ist.
<br>
<br>
Zum Beispiel bedeutet $\frac{3}{4}$, dass drei von vier gleichen Teilen eines ganzen betrachtet wird.

Das Rechnen mit Brüchen ist wichtig, da es häufig notwendig ist, Brüche zu addieren, zu subtrahieren, zu multiplizieren oder zu dividieren, um Daten oder Anteile zu berechnen.

**Addition und Subtraktion:**

Um Brüche zu addieren oder zu subtrahieren, müssen sie den gleichen Nenner haben. Wenn dies nicht der Fall ist, kann der gemeinsame Nenner gefunden werden, indem die Nenner der beiden Brüche multipliziert werden.

$$\frac{a}{b} + \frac{c}{d} = \frac{c \cdot d + c \cdot b}{b \cdot d}$$

**Multiplikation:**

Zwei Brüche werden multipliziert, indem die Zähler und die Nenner miteinander multipliziert werden.

$$\frac{a}{b} \cdot \frac{c}{d} = \frac{a \cdot c}{b \cdot d}$$

**Division:**

Ein Bruch wird durch einen anderen Bruch dividiert, indem der erste Bruch mit dem Kehrwert (Umkehrung) des zweiten Bruchs multipliziert wird.

$$\frac{a}{b} \div \frac{d}{d} = \frac{a \cdot d}{b \cdot c}$$

Brüche kann man auch als Dezimalzahlen, auch Kommazahlen genannt, darstellen. So eine Dezimalzahl besteht immer aus den Vorkommastellen, dem Komma und den Nachkommastellen.
Einige Beispiele sind:
- $\frac{75}{100}=0.75$
- $\frac{1}{100}=0.01$
- $\frac{1}{4} = 0.25$
- $\frac{1}{3} = 0.333...$
- $\pi = 3.14159...$
- $g = 9.81...$
