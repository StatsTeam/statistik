## 2. Aufgabe

Ein Unternehmen hat $374$ Angestellte, deren jährliche Gehälter in Euro erfasst wurden. Du möchtest ein Histogramm erstellen, um die Gehaltsverteilung darzustellen. Bestimme die Anzahl der Klassen für das Histogramm mithilfe der Sturges' Regel.

### Lösung

$k=1+log_2n=1+log_2{(374)}\approx{9,54}\approx{10}$ Klassen

## 3. Aufgabe

Eine Umfrage wurde unter $250$ Teilnehmern durchgeführt, und du möchtest die Verteilung der Antworten in einem Histogramm darstellen. Verwende die Wurzelformel, um die Anzahl der Klassen zu berechnen.

### Lösung

$k\approx\sqrt{n}\approx{\sqrt{250}}\approx{15,81}\approx{16}$ 

15 Klassen wären aber auch eine gute Wahl.

## 4. Aufgabe

Angenommen, du hast eine Stichprobe von $150$ Datenpunkten. Der kleinste Wert in deinem Datensatz ist $10$, und der größte Wert ist $70$. Berechne die Klassenbreite, wenn du $10$ Klassen verwenden möchtest.

### Lösung

$\text{Klassenbreite}=\frac{70-10}{10}=6$

## 5. Aufgabe

Ein Datensatz umfasst 400 Einträge zur monatlichen Arbeitszeit von Angestellten. Berechne die Anzahl der Klassen, die du mit der Sturges' Regel und der Wurzelformel erhältst. Vergleiche die Ergebnisse und entscheide, welche Klassenanzahl sinnvoller erscheint.

### Lösung

**Sturges' Regel:**

$k=1+log_2n=k=1+log_2{(400)}\approx{9,64}\approx{10}$

**Wurzel Formel:**

$k\approx\sqrt{n}\approx{\sqrt{400}}=20$

Welche Klassenanzahl ist sinnvoller?
- **Sturges' Regel** führt zu einer geringeren Anzahl von Klassen (10), was zu einer einfacheren Darstellung der Daten führt. Diese Regel eignet sich gut, wenn man eine relativ grobe, übersichtliche Darstellung bevorzugt und davon ausgeht, dass die Daten einer normalverteilten Struktur folgen.
</br>
- Die **Wurzelformel** schlägt eine größere Klassenanzahl (20) vor, was eine feinere Aufteilung der Daten ermöglicht. Diese Methode ist sinnvoller, wenn man die Daten genauer aufschlüsseln möchte, insbesondere wenn es in den Daten viele Nuancen oder Ausreißer gibt, die durch mehr Klassen besser erfasst werden können.

## 6. Aufgabe

Ein Datensatz umfasst 80 Messwerte eines Experiments, bei dem die Temperatur in Grad Celsius gemessen wurde. Die Werte liegen zwischen 15°C und 45°C. Du möchtest ein Histogramm mit etwa 8 Klassen erstellen. Berechne die Klassenbreite, und diskutiere, ob es sinnvoll wäre, eine etwas abweichende Klassenbreite zu wählen, um die Daten besser darzustellen.

### Lösung

$\text{Klassenbreite}=\frac{45-15}{8}={3,75}$

**Diskussion:**

Die Klassenbreite von 3,75 ist mathematisch korrekt. Es könnte jedoch sinnvoll sein, eine leicht abweichende Klassenbreite zu wählen, um eine bessere Lesbarkeit und eine übersichtlichere Einteilung zu erhalten. Eine Breite von 3,75°C ist nicht einfach zu interpretieren, und es könnte kompliziert sein, die Klassengrenzen zu wählen und darzustellen.
<br>
<br>
Man könnte die Klassenbreite auf 5°C aufrunden, da dies eine einfacher handhabbare Zahl ist. Dadurch hätte man gleichmäßigere, besser interpretierbare Klassengrenzen, wie etwa 15-20°C, 20-25°C usw.
<br>
<br>
Mit einer gerundeten Klassenbreite von 5°C würden sich dann 9 Klassen ergeben.
<br>
<br>
Man könnte also entweder 8 oder 9 Klassen verwenden, je nach Präferenz und Detailgenauigkeit.
<br>
<br>
Eine runde Klassenbreite wie 5°C würde die Darstellung und Interpretation erleichtern, ohne die Daten zu stark zu verfälschen.

## 7. Aufgabe

Du hast einen kleinen Datensatz mit nur 20 Werten, die du analysieren möchtest. Berechne die Anzahl der Klassen mit der Sturges' Regel und der Wurzelformel. Diskutiere, welche Methode bei so kleinen Datensätzen sinnvoller ist und warum.

### Lösung

Sturges' Regel:<br>
$k=1+log_2(n)=k=1+log_2{(20)}\approx{5,32}\approx{6}$

Wurzel Formel:<br>
$k\approx\sqrt{20}\approx{\sqrt{4,47}}=5$

Bei kleinen Datensätzen, wie in diesem Beispiel mit nur 20 Werten, ist es oft besser, weniger Klassen zu verwenden, da zu viele Klassen dazu führen können, dass jede Klasse nur sehr wenige Werte enthält, was das Histogramm weniger aussagekräftig macht.

## 8. Aufgabe

Ein Datensatz enthält 5000 Datenpunkte zu den Verkaufszahlen eines Unternehmens. Die Werte reichen von 100 bis 2000 Einheiten. Verwende die Wurzelformel, um die Anzahl der Klassen zu berechnen, und bestimme dann die Klassenbreite. Diskutiere, ob es sinnvoll wäre, die Anzahl der Klassen anzupassen, um eine präzisere Darstellung zu ermöglichen.

### Lösung

$k\approx\sqrt{5000}\approx70,71\approx71$

$\text{Klassenbreite}=\frac{2000-100}{71}={26,76}\approx27$

In dieser Aufgabe führt die Wurzelformel zu einer sehr hohen Anzahl von Klassen (71), was möglicherweise zu einer übermäßig detaillierten und unübersichtlichen Darstellung führt. Eine so feine Aufteilung der Daten könnte sinnvoll sein, wenn die Daten sehr gleichmäßig verteilt sind und eine präzise Analyse erforderlich ist. Sollte die Verteilung der Daten jedoch stark gruppiert sein oder sich in bestimmten Bereichen häufen, wäre es ratsam, die Anzahl der Klassen zu verringern, um eine klarere Übersicht zu erhalten.
<br>
<br>
In vielen Fällen könnte es hilfreich sein, die Anzahl der Klassen auf etwa 50 oder weniger zu reduzieren, um eine aussagekräftigere Darstellung zu gewährleisten. Mit weniger Klassen lassen sich Muster oder Verteilungen in den Daten oft klarer erkennen, ohne dass die Darstellung durch zu viele Unterteilungen überfrachtet wird. Eine Alternative wäre, die Klassenbreite geringfügig anzupassen, beispielsweise auf 30 statt 27 Einheiten, was eine bessere Anpassung der Einteilung an die Datenstruktur ermöglichen würde. Dies würde die Interpretation erleichtern, ohne dabei die Genauigkeit der Analyse zu beeinträchtigen.

## 9. Aufgabe

Du hast folgende Daten, welche die Geschwindigkeiten von 56 Autos auf einer Autobahn (in km/h) darstellen:

$81, 84, 91, 92, 94, 99, 101, 106, 109, 112, 115, 119, 123, 125, 127, 131, 135, 138, 139, 142, 145, 145, 148, 151, 155, 159, 159,\\ 164, 167, 169, 172, 176, 181, 184, 188, 191, 196, 199, 204, 207, 211, 215, 218, 223, 227, 231, 235, 239, 243, 248, 252, 257, 259, 264, 268, 273$

1. Berechne zunächst die Anzahl der Klassen mithilfe der Sturges' Regel und der Wurzelformel.
2. Bestimme für beide Methoden die passende Klassenbreite.
3. Erstelle je ein Histogramm für die Daten auf Basis der berechneten Klassenanzahl für Sturges' Regel und für die Wurzelformel.
4. Vergleiche die beiden Histogramme und diskutiere, wie sich die Wahl der Klassenanzahl auf die Darstellung der Daten auswirkt. Welches Histogramm bietet eine bessere Übersicht und Analyse?

### Lösung

1. Berechne zunächst die Anzahl der Klassen mithilfe der Sturges' Regel und der Wurzelformel.
   - Wurzel Formel: $k\approx\sqrt{56}\approx7,48\approx8$
   - Sturges' Regel: $k=1+log_2(n)=k=1+log_2{(56)}\approx{6,8}\approx{7}$

2. Bestimme für beide Methoden die passende Klassenbreite.
   - Klassenbreite bei Wurzel Formel: $\text{Klassenbreite}=\frac{273-81}{8}={24}$
   - Klassenbreite bei Sturges' Regel: $\text{Klassenbreite}=\frac{273-81}{7}={27,43}\approx28$

4. Vergleiche die beiden Histogramme und diskutiere, wie sich die Wahl der Klassenanzahl auf die Darstellung der Daten auswirkt. Welches Histogramm bietet eine bessere Übersicht und Analyse?
   
   - **Histogramm basierend auf der Sturges' Regel:**
      - Die Sturges' Regel hat eine Klassenanzahl von 7 mit einer Klassenbreite von ca. 28 km/h.
      - Dieses Histogramm zeigt eine grobe Unterteilung der Daten. Es verdichtet die Geschwindigkeiten in größere Gruppen, was zu einem weniger detaillierten Überblick führt.
      - Das Histogramm ist nützlich, wenn man eine schnelle und übersichtliche Analyse der Verteilung sucht, aber es könnte einige interessante Feinheiten der Datenstruktur verschleiern. Große Schwankungen oder kleinere Muster in der Verteilung der Geschwindigkeiten könnten hier untergehen.
      
   - **Histogramm basierend auf der Wurzelformel:** 
      - Die Wurzelformel führt zu einer feineren Einteilung mit 8 Klassen und einer Klassenbreite von 24 km/h.
      - Im Vergleich zum Sturges-Histogramm bietet dieses eine etwas detailliertere Ansicht der Verteilung. Es zeigt mehr Nuancen in der Verteilung der Geschwindigkeiten und bietet potenziell mehr Informationen für eine genauere Analyse.
      - Bei der Anwendung der Wurzelformel ist das Histogramm jedoch manchmal zu detailliert und könnte unnötige Komplexität einführen, insbesondere bei kleineren Datensätzen oder wenn es nur wenige starke Muster in den Daten gibt.
