# Das geometrische Mittel

Diese Lagemaßzahl wird zur Berechnung der durchschnittlichen Veränderungsrate verwendet, insbesondere wenn die Daten in Form von Verhältnissen oder Wachstumsraten vorliegen. Einfach gesagt bedeutet das, dass man mit dem geometrischen Mittel zum Beispiel die  durchschnittliche prozentuale Veränderungen von Wachstumsraten bestimmen kann. Außerdem ist diese Lagemaßzahl stets kleiner oder gleich dem arithmetischen Mittel.

## Einführungsbeispiel

Eine Firma hat im letzten Jahr ihren Umsatz verdoppelt, dieses Jahr war der Umsatz sogar achtmal so groß wie im Jahr zuvor. Wie groß ist die durchschnittliche Steigerung des Umsatzes in den beiden letzten Jahren?
<br>
<br>
Diese Informationen tabellarisch visualisiert können den Sachverhalt verständlicher darstellen:

| Jahr | Umsatz                         |
| ---- | ------------------------------ |
| 1.   | $x$                            |
| 2.   | $2\cdot{x}$                    |
| 3.   | $8\cdot{2}\cdot{x}=16\cdot{x}$ |

Dabei ist $x$ der Umsatz im 1. Jahr. Grafisch könnte man den Sachverhalt folgendermaßen darstellen:

<img src="../img/geometrisches_mitt.svg" alt="Empirische Verteilungsfunktion" width="300">

Würde man das arithmetische Mittel nun verwenden, um die durchschnittliche Steigerung der beiden letzten Jahre (2. Jahr und 3. Jahr) zu berechnen, so würde man auf $(2+8)/2=5$ kommen.
Mit der durchschnittlichen Steigerung von $5$ stimmt jedoch die Berechnung in der Tabelle nicht mehr. Denn es würde zu folgenden Werten führen:

| Jahr | Umsatz                         |
| ---- | ------------------------------ |
| 1.   | $x$                            |
| 2.   | $5\cdot{x}$                    |
| 3.   | $5\cdot{5}\cdot{x}=25\cdot{x}$ |


Wir sehen deutlich dass wir nicht auf $16$ kommen sondern auf $5\cdot{5}=25$, somit kann die $5$ nicht die durchschnittliche Steigerung sein!
Nun ist klar dass man in diesem Beispiel nicht das arithmetische Mittel verwenden darf.
<br>
<br>
Aus diesem Beispiel geht eine wichtige Information hervor, wir suchen eine Zahl, welche mit sich selber multipliziert die Zahl $16$ ergibt. Dabei kann es nur die $\sqrt{2\cdot{8}}=\sqrt{16}=4$ sein. Der auf diese Weise gebildete Mittelwert heißt das geometrische Mittel der Zahlen $2$ und $8$.

## Definition

> Sind $x_1$, $x_2$, $...,$ $x_n$ reelle Zahlen mit der Eigenschaft $x_i\geq{0}$, $i=1$, $2$, $...,$ $n$ 
> dann heißt: $$\overline{x_g}=\sqrt[n]{x_1\cdot{x_2}\cdot{}...\cdot{x_n}}=\sqrt[n]{\prod_{i=1}^nx_i}$$
> das geometrische Mittel dieser Zahlen.

## Geometrische Interpretation

Das geometrische Mittel der beiden Zahlen $2$ und $8$ kann einfach veranschaulicht werden. Ein Rechteck mit den Seitenlängen $2$ und $8$ soll in ein flächengleiches Quadrat umgewandelt werden. Dann müssen die Seitenlängen des Quadrats gleich der Wurzel aus $16$ sein, denn 16 ist der Flächeninhalt des Rechtecks:

<img src="../img/rechteck.drawio 1.svg" alt="Empirische Verteilungsfunktion" width="300">

Auch bei einem Quader mit den Seitenlängen von z.B. $2$, $3$ und $4$ ist das Prinzip identisch. Hier wird ein Würfel gesucht dessen Volumen dem Volumen des Quaders entspricht und alle drei Seitenlängen gleich sind:
$Seitenlänge=\sqrt[3]{2\cdot{3}\cdot{4}}=\sqrt[3]{24}$

## Anwendung

Im Gegensatz zum arithmetischen Mittel, das die Summe der Werte durch ihre Anzahl teilt, berücksichtigt das geometrische Mittel das Verhältnis zwischen den Werten.
<br>
<br>
Das geometrische Mittel wird in verschiedenen Bereichen angewendet, wie zum Beispiel in der Finanzwelt zur Berechnung von Renditen über mehrere Jahre, in der Biologie zur Analyse von Wachstumsraten oder in der Technik zur Berechnung von Effizienzsteigerungen.

## 1. Anwendungsbeispiel

Das beste Beispiel sind die von den Banken angebotene Sparpläne, bei denen sich der Zinssatz von Jahr zu Jahr ändert. Für den Sparer ist es dabei interessant, festzustellen, wie groß der durchschnittliche Zinssatz über die gesamte Laufzeit ist. Angenommen eine Bank bietet einen Sparplan über zwei Jahre an. Im 1. Jahr beträgt der Zinssatz $p_1=1\%$, im 2. Jahr $p_2=2\%$. Wie groß ist der durchschnittliche Zinssatz?
<br>
<br>
Das Kapital wächst im 1. Jahr auf das 1,01-fache und im 2. Jahr auf das 
1,02-fache. Aus dieser Information folgt für das geometrische Mittel: $\overline{x_g}=\sqrt{1.01\cdot{1.02}}\approx{1.0149}$
Somit ist der durchschnittliche Zinssatz etwa $1.49\%$ groß und 
nicht $1.5\%$ (wie man beim flüchtigen Hinsehen glauben mag).

## 2. Anwendungsbeispiel

Ein Guthaben $W$ wurde in den letzten Jahren wie folgt verzinst:

| Jahr | Zinssatz |
| ---- | -------- |
| 2007 | $+12\%$  |
| 2008 | $+7\%$   |
| 2009 | $+2\%$   |
| 2010 | $-4\%$   |
| 2011 | $-10\%$  |

Nun soll der durchschnittliche, also konstante, Zinssatz $p$ ermittelt werden.

Der Zuwachs im 1. Jahr ist nach der Zinsrechnung:<br>
$W_1=W+W\cdot{0.12}=(1+0.12)\cdot{W}=1.12\cdot{W}$
<br>
<br>
Das Gleich gilt auch für das zweite Jahr, nur muss jetzt beachtet werden, dass zum Grundbetrag $W$ schon Geld dazugekommen ist, und man nun den Wert $W_1$ verzinst:<br>
$W_2=W_1+W_1\cdot{0.07}=(1+0.07)\cdot{W_1}=1.07\cdot{W_1}=1.07\cdot{1.12}\cdot{W}$
<br>
<br>
Man sieht, dass die Zinsfaktoren einfach multipliziert werden müssen, damit man den Endbetrag $W_{Ende}$ erhält, also folgt damit:<br>
$W_{Ende}=W\cdot{1.12}\cdot{1.07}\cdot{1.02}\cdot{0.96}\cdot{0.90}$
<br>
<br>
Nun Stellt man sich die Frage, welche Zahl muss man fünfmal mit sich selbst multiplizieren, um den gleichen Wert zu erhalten, den die fünf Faktoren ergeben:<br>
$p=\sqrt[5]{1.12\cdot{1.07}\cdot{1.02}\cdot{0.96}\cdot{0.90}}\approx{1.01}$
<br>
<br>
Genau das ist wieder das geometrische Mittel $\overline{x_g}$.
