## Aufgabe 1

### Gib die folgenden mengen im aufzählenden Verfahren an:

- $M_1 = \{ x \in \mathbb{N} \mid 3 \leq x \leq 8\}$

- $M_2 = \{ y \in \mathbb{Z} \mid -2 < y \leq 4\}$

- $M_3 = \{ s \in \mathbb{P} \mid s \leq 27\}$

- $M_4 = \{ z \in \mathbb{Z} \mid |z| < 5 \}$
  - Bei $|z|$ die beiden Striche, sind sogenannte Betragsstriche. Sie bedeuten dass die Zahl $z$ immer größer $0$ sein soll.

### Lösungen Aufgabe 1

| Menge | Lösung                       |
| ----- | ---------------------------- |
| $M_1$ | $\{3,4,5,6,7,8\}$            |
| $M_2$ | $\{-1,0,1,2,3,4\}$           |
| $M_3$ | $\{2,3,5,7,11,13,17,19,23\}$ |
| $M_4$ | $\{1,2,3,4\}$                |

## Aufgabe 3

## Aufgabe 6 

### Gib an, welche der folgenden Aussagen richtig bzw. falsch sind!


(a) Beim Addieren von natürlichen Zahlen $\{ x \in \mathbb{N}\}$ entsteht stets wieder eine natürliche Zahl  $\{ y \in \mathbb{N}\}$.

(b) Beim Subtrahieren von natürlichen Zahlen $\{ x \in \mathbb{N}\}$ entsteht immer eine rationale Zahl $\{ y \in \mathbb{Q}\}$.

(c) Beim dividieren ganzer Zahlen $\{ x \in \mathbb{Z}\}$ entsteht stest eine ganze Zahl $\{ y \in \mathbb{Z}\}$

(d) Jede reelle Zahl $\{ x \in \mathbb{R}\}$ ist eine rationale Zahl $\{ x \in \mathbb{Q}\}$.

(e) Jede rationale Zahl $\{ x \in \mathbb{Q}\}$ ist eine reelle Zahl $\{ x \in \mathbb{R}\}$.

(f) Jede ganze Zahl $\{ x \in \mathbb{Z}\}$ ist eine natürliche Zahl $\{ x \in \mathbb{N}\}$.

(g) Beim Dividieren von natürlichen Zahlen $\{ x \in \mathbb{N}\}$ entsteht immer eine reelle Zahl $\{ y \in \mathbb{R}\}$ .
