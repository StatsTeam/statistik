## 3. Aufhabe

Ein Landwirt weiß, dass die Getreidesorte, die er letztes Jahr gepflanzt hatte im Durchschnitt 25 Körner pro Ähre erbrachte. Eine Ähr ist der Teil einer Getreidepflanze, an dem die Körner wachsen.
<br>
<br>
Dieses Jahr pflanzte er eine andere Sorte und betrachtet nun eine Stichprobe von 16 Kornähren. Er zählt die Körner pro Ähre und erhält folgende Stichprobe:

| Index     | **1**  | **2**  | **3**  | **4**  | **5**  | **6**  | **7**  | **8** | **9** |
| --------- | ------ | ------ | ------ | ------ | ------ | ------ | ------ | ----- | ----- |
| Anzahl    | 41     | 36     | 25     | 36     | 41     | 32     | 41     | 36    | 32    |
| **Index** | **10** | **11** | **12** | **13** | **14** | **15** | **16** |       |       |
| Anzahl    | 25     | 36     | 36     | 41     | 25     | 36     | 25     |       |       |

Geben Sie die empirische Verteilungsfunktion der Körner pro Ähre an.

### Lösung

Als erstes wird die Anzahl bzw. die absolute Häufigkeit der einzelnen Merkmalsausprägungen bestimmt:

| Index $i$ | **Anzahl** | Absolute Häufigkeit $n_i$ |
| --------- | ---------- | ------------------------- |
| 1         | 25         | 4                         |
| 2         | 32         | 2                         |
| 3         | 36         | 6                         |
| 4         | 41         | 4                         |

Dann werden die relativen Häufigkeiten berechnet:

$$n=16$$

$$h_i=\frac{n_i}{n}$$

$$h_1=\frac{4}{16}=0,25$$

$$h_2=\frac{2}{16}=0,125$$

$$h_3=\frac{6}{16}=0,375$$

$$h_4=\frac{4}{16}=0,25$$

| Index $i$ | **Anzahl** | Absolute Häufigkeit $n_i$ | Relative Häufigkeit $h_i$ |
| --------- | ---------- | ------------------------- | ------------------------- |
| 1         | 25         | 4                         | 0,25                      |
| 2         | 32         | 2                         | 0,125                     |
| 3         | 36         | 6                         | 0,375                     |
| 4         | 41         | 4                         | 0,25                      |

Jetzt kann man die kumulierten Häufigkeiten berechnen:

$$F(x_i)=h_1+h_2+...h_i=\sum_{j=1}^{i}h_j$$

$$F(x_1)=h_1=0,25$$

$$F(x_2)=h_1+h_2=0,25+0,125=0,375$$

$$F(x_3)=h_1+h_2+h_3=0,25+0,125+0,375=0,75$$

$$F(x_4)=h_1+h_2+h_3+h_4=0,25+0,125+0,375+0,25=1$$

| Index $i$ | **Anzahl** | Absolute Häufigkeit $n_i$ | Relative Häufigkeit $h_i$ | Kumulierte Häufigkeit $F(x)$ |
| --------- | ---------- | ------------------------- | ------------------------- | ---------------------------- |
| 1         | 25         | 4                         | 0,25                      | 0,25                         |
| 2         | 32         | 2                         | 0,125                     | 0,375                        |
| 3         | 36         | 6                         | 0,375                     | 0,75                         |
| 4         | 41         | 4                         | 0,25                      | 1                            |

Aufstellen der empirischen Verteilungsfunktion:

$$F(x) =
\begin{cases}
0 & \text{für } x < 25 \\
0{,}25 & \text{für } 25 \leq x < 32 \\
0{,}375 & \text{für } 32 \leq x < 36 \\
0{,}75 & \text{für } 36 \leq x < 41 \\
1 & \text{für } x \geq 41
\end{cases}$$
