# Aufgaben zu: Quantile und Boxplots

## 1. Aufgabe

Berechnen Sie das $0,25$-Quantil mit den folgenden Daten:
<br>
`6,58; 6,61; 6,63; 6,69; 6,74; 6,79; 6,81; 6,95; 7,08; 7,15; 7,17`

### Lösung

Gesamztahl der Daten:<br>
$n=11$

$0,25$-Quantil:<br>
$p=0.25$

$n\cdot{p}=11\cdot0.25=2.75$ ist keine ganze Zahl, somit wird die Zahl abgerundet auf $2$!

$x_{0.25}=x_{2+1}=x_3=6.63$
