# Aufgaben zu: Das geometrische Mittel

## 1. Aufgabe

Sie haben im Jahr 2010 Aktien eines Unternehmens gekauft. Seitdem hat die Aktie folgende Wertveränderungen erfahren:

| Jahr | Zinssatz |
| ---- | -------- |
| 2010 | $+15\%$  |
| 2011 | $+8\%$   |
| 2012 | $-3\%$   |
| 2013 | $+20\%$  |
| 2014 | $-10\%$  |
| 2015 | $+13\%$  |
| 2016 | $+25\%$  |

Wie hoch ist die durchschnittliche Wertveränderung?

### Lösung

$$\sqrt[7]{(1+0.15)\cdot(1+0.08)\cdot(1-0.03)\cdot(1+0.2)\cdot(1-0.1)\cdot(1+0.13)\cdot(1+0.25)}=1.0908\approx109.08\%$$

$$109.08\% - 100\%=9.08\%$$

## 2. Aufgabe

Man möchte ein Sparguthaben in Höhe von 1000€ in ein Projekt investieren. Dieses Projekt hat eine Laufzeit von fünf Jahren und weist folgende jährliche Verzinsung auf:

| Zeitpunkt | Zinssatz $i$ in % | Jahresendwert in € |
| --------- | ----------------- | ------------------ |
| $t_0=0$   | -                 | 1000               |
| $t_1=1$   | 10                | 1100               |
| $t_2=2$   | $i_2$             | 1232               |
| $t_3=3$   | 35                | 1663.20            |
| $t_4=4$   | 27                | 2112.26            |
| $t_5=5$   | 4                 | 2196.75            |

1. Zeigen Sie, dass der Zinssatz $i_2$ für den Zeitpunkt $t_2$ bei exakt 12% bzw. der zugehörige Wachstumsfaktor $x_2$ bei $1.12$ liegt!
2. Zeigen Sie, dass der auf Basis des arithmetischen Mittels ermittelte durchschnittliche Wachstumsfaktor pro Jahr bei $1.12$ liegt!
3. Überprüfen die Ihr Ergebnis aus Aufgabe zwei, indem Sie das Startkapital mit dem ermittelten Durchschnittsfaktors über 5 Jahre verzinsen! Was fällt Ihnen auf?
4. Berechnen Sie nun den korrekten durchschnittlichen Wachstumsfaktor pro Jahr!
5. Überprüfen Sie ihr Ergebnis aus der vierten Aufgabe, indem Sie über 5 Jahre das Startkapital verzinsen!

### Lösung

1. Zeigen Sie, dass der Zinssatz $i_2$ für den Zeitpunkt $t_2$ bei exakt 12% bzw. der zugehörige Wachstumsfaktor $x_2$ bei $1.12$ liegt!

    Anwendung des Dreisatzes: 

    $$\frac{100}{1100}=\frac{i_2}{1232}$$

    $$i_2=\frac{10\cdot1232}{1100}=112\%$$

    $$112\%-100\%=12\%$$

    $$x_2=1.12-1=0.12$$

2. Zeigen Sie, dass der auf Basis des arithmetischen Mittels ermittelte durchschnittliche Wachstumsfaktor pro Jahr bei $1.176$ liegt!

    $$\overline{x}=\frac{1}{5}(1.1+1.12+1.35+1.27+1.04)=1.176$$

3. Überprüfen die Ihr Ergebnis aus Aufgabe zwei, indem Sie das Startkapital mit dem ermittelten Durchschnittsfaktors über 5 Jahre verzinsen! Was fällt Ihnen auf?

    $$1000€ \cdot{1.176}\cdot{1.176}\cdot{1.176}\cdot{1.176}\cdot{1.176}=2249.24€$$

    $$2249.24€\neq2196.75€$$

    Laut der Tabelle müsste man auf $2196.75€$ kommen, was nicht der Fall ist. Das arithmetische Mittel ist falsch!

4. Berechnen Sie nun den korrekten durchschnittlichen Wachstumsfaktor pro Jahr!

    $$\overline{x_g}=\sqrt[5]{1.1\cdot1.12\cdot1.35\cdot1.27\cdot1.04}=1.170459$$

5. Überprüfen Sie ihr Ergebnis aus der vierten Aufgabe, indem Sie über 5 Jahre das Startkapital verzinsen!

    $$1000€ \cdot{1.170459}\cdot{1.170459}\cdot{1.170459}\cdot{1.170459}\cdot{1.170459}=2196.75€$$

## 3. Aufgabe

Jemand investiert sein gesamtes Sparguthaben (in beliebiger Höhe) in Aktien, die er über drei Jahre hält und beobachtet. Die jeweiligen  Monatsendwerte der  Aktien fasst er wie folgt zusammen:

| Zinssatz $i$ in % | Absolute Häufigkeit |
| ----------------- | ------------------- |
| -20               | 1                   |
| -5                | 3                   |
| 0                 | 6                   |
| +5                | 20                  |
| +15               | 4                   |
| +50               | 2                   |

1. Bestimmen Sie den jeweiligen zu seinem Zinssatz i gehörenden Wachstumsfaktor!
2. Zeigen  Sie,  dass  der  auf  Basis  des  arithmetischen  Mittelwerts  berechnete durchschnittliche Wachstumsfaktor pro Jahr bei rund 1,0625 liegt!
3. Zeigen Sie nun, dass der korrekte durchschnittliche Wachstumsfaktor pro Jahr bei rund 1,056211 liegt!

### Lösung

1. Bestimmen Sie den jeweiligen zu seinem Zinssatz i gehörenden Wachstumsfaktor!

    - $a_1 = 1-0.2=0.8$
    - $a_2 = 1-0.05=0.95$
    - $a_3 = 1$
    - $a_4 = 1+0.05=1.05$
    - $a_5 = 1+0.15=1.15$
    - $a_6 = 1+0.5=1.5$

2. Zeigen  Sie,  dass  der  auf  Basis  des  arithmetischen  Mittelwerts  berechnete durchschnittliche Wachstumsfaktor pro Jahr bei rund 1,0625 liegt!

$$\overline{x}=\frac{0.8\cdot1+0.95\cdot3+1\cdot6+1.05\cdot20+1.15\cdot4 + 1.5\cdot2}{36}=1.0625$$

3. Zeigen Sie nun, dass der korrekte durchschnittliche Wachstumsfaktor pro Jahr bei rund 1,056211 liegt!

$$\overline{x_g}=\sqrt[36]{0.8^1\cdot0.95^3\cdot1^6\cdot1.05^{20}\cdot1.15^4\cdot1.5^2}=1.056211$$
