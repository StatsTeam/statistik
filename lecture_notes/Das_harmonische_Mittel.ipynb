# Das harmonische Mittel

Das harmonische Mittel ist ebenfalls eine Lagemaßzahl, welche den Durchschnitt positiver Zahlen angibt. Im Gegensatz zum arithmetischen Mittel, bei dem die Zahlen addiert und durch die Anzahl der Zahlen geteilt werden, verwendet das harmonische Mittel die Kehrwerte der Zahlen.
Es wird oft verwendet, um eine durchschnittliche Rate oder ein Durchschnittsverhältnis zu bestimmen.

## Einführungsbeispiel

Ein Rennfahrer soll zwei Runden auf einer Rennstrecke fahren. Die erste Runde fährt er mit einer Geschwindigkeit von $v_1=50\ km/h$. Wie schnell muss er die zweite Runde fahren, damit sich für beide eine Durchschnittsgeschwindigkeit von $80\ km/h$ ergibt?
<br>
<br>
Ein häufige Antwort auf diese Fragestellung lautet:<br>
$v_2=110\ km/h$ wegen $\frac{50+110}{2}=80\ km/h$.

Jedoch vergisst man dabei, dass für die erste Runde der Fahrer mehr Zeit benötigt hatte als für die zweite Runde. Denn die erste Runde wurde mit einer geringeren Geschwindigkeit durchlaufen als die zweite Runde.
Genau aus diesem Grund kann man auch das arithmetische Mittel nicht verwenden! 
<br>
<br>
Wenn man z.B. eine Stunde lang mit einer Geschwindigkeit von $60\ km/h$ fährt und anschließend eine Minute mit $40\ km/h$, würde kein vernünftiger Mensch auf die Idee kommen, dass man mit einer Durchschnittsgeschwindigkeit von $\frac{60+40}{2}=50\ km/h$ fährt. 
Dieses Gedankenexperiment sollte nochmals verdeutlichen, dass das arithmetische Mittel hier nicht verwendet werden kann.
<br>
<br>
Für Probleme dieser Art bestimmt man den gesamten gefahrenen Weg und teilt diesen durch die dafür benötigte Zeit. Genau dass ist dann die Durchschnittsgeschwindigkeit.
<br>
<br>
Der Rennfahrer fährt in der ersten Runde die Strecke $s$ und in der zweiten ebenfalls. Deswegen ist die Gesamte gefahrene Strecke $2s$.
Für die erste Runde benötigt der Rennfahrer die Zeit $t_1=s/v_1$
und für die zweite Runde die Zeit $t_2=s/v_2$.
Somit ist die Durchschnittsgeschwindigkeit des Rennfahrers:

$$v=\frac{2s}{s/v_1+s/v_2}=\frac{2}{1/v_1+1/v_2}$$

Einsetzen der gegebenen Zahlenwerte in die Gleichung ergibt:

$$80=\frac{2}{1/50+1/v_2}$$

Nun kann man diese Gleichung nach der gesuchten Größe $v_2$ umstellen:

$$v_2=200\ km/h$$

Somit ist die Geschwindigkeit in der 2. Runde $200\ km/h$ groß und nicht $80\ km/h$.

Betrachten wir einen etwas anderen allgemeineren Ansatz. Dafür wird kein Zahlenwert für die Durchschnittsgeschwindigkeit $v$ festgelegt und wir arbeiten mit der folgenden Gleichung: 

$$v=\frac{2}{1/50+1/v_2}$$

Umgestellt nach $v_2$ ergibt sich: 

$$v_2(v)=\frac{50\cdot{v}}{100-v}$$

An dem Graphen kann man nun ablesen welchen Wert $v_2$ haben muss, wenn eine Durchschnittsgeschwindigkeit $v$ vorgegeben ist.
<br>
<br>
Die wichtigste Schlussfolgerung aus den ganzen Betrachtungen ist: 
Die Durchschnittsgeschwindigkeit $v=\frac{2}{\frac{1}{v_1}+\frac{1}{v_2}}$ aus den einzelnen Werten $v_1$ und $v_2$ nennt man das harmonische Mittel aus $v_1$ und $v_2$.

## Definition

> Es seien $x_1$, $x_2$, $...,$ $x_n$ reelle Zahlen, die alle größer als null sind. 
> Dann heißt: 
> $$\overline{x_H}=\frac{n}{\frac{1}{x_1}+\frac{1}{x_2}+\ ...\ +\frac{1}{x_n}}=\frac{n}{\sum_{i=1}^n\frac{1}{x_i}}$$
> das harmonische Mittel dieser Zahlen.

## 1. Anwendungsbeispiel

Es soll das harmonische Mittel der Zahlenwerte $1$, $2$ und $3$ berechnet werden.

**Lösung:**

$\overline{x_H}=\frac{3}{\frac{1}{1}+\frac{1}{2}+\frac{1}{3}}\approx{1,64}$

## 2. Anwendungsbeispiel

Es liegen die Daten der Heizölpreise von Deutschland im Jahre 2012 bis 2017 vor:

| Jahr | Preis in Cent pro Liter |
| ---- | ----------------------- |
| 2012 | 88,1                    |
| 2013 | 82,9                    |
| 2014 | 76,4                    |
| 2015 | 58,8                    |
| 2016 | 48,85                   |
| 2017 | 59,02                   |

Ein Hausbesitzer kauft in diesem Zeitraum Heizöl für seine Heizung. 
Wie groß ist der durchschnittliche Literpreis von 2012 bis 2017, wenn
1. jedes Jahr die gleiche Menge Heizöl gekauft wird?
2. jedes Jahr derselbe Geldbetrag für Heizöl ausgegeben wird?

**Lösung:**

1.  $$\overline{x}=\frac{88,1+82,9+76,4+58,8+48,85+59,02}{6}=69,01\ Cent\ Pro\ Liter$$
2.  $$\overline{x_H}=\frac{6}{\frac{1}{88,1}+\frac{1}{82,9}+\frac{1}{76,4}+\frac{1}{58,8}+\frac{1}{48,85}+\frac{1}{59,02}}\approx{65,99}$$

## 3. Anwendungsbeispiel

Student Sauerbrot ist der Meinung, dass ihm das Studentenleben zu gut bekommt und
möchte etwas gegen seinen deutlich sichtbaren Bauch tun. Zu diesem Zweck beschließt
er, jeden Tag eine Stunde Fahrrad zu fahren. In der ersten Woche schafft er folgende
Strecken (in km):
<br>
<br>
`15; 16,5; 17,5; 18; 18; 20; 22`

1. Sauerbrot möchte wissen, welche Geschwindigkeit er im Schnitt geschafft hat und
berechnet das arithmetische Mittel (Wert?)
2. Sein Bruder glaubt dagegen, man müsse bei einer solchen Fragestellung das harmonische Mittel berechnen (Wert?).
3.  Wer von beiden hat Recht?
4.  Um wie viel Prozent hat Sauerbrot seine Kilometerleistung im Schnitt gesteigert?

### Lösung

1. Sauerbrot möchte wissen, welche Geschwindigkeit er im Schnitt geschafft hat und
berechnet das arithmetische Mittel (Wert?)

 $$\overline{x}=\frac{15+16.5+17.5+18+18+20+22}{7}=18.1429$$

2. Sein Bruder glaubt dagegen, man müsse bei einer solchen Fragestellung das harmonische Mittel berechnen (Wert?).

    $$\overline{x_H}=\frac{7}{15^{-1}+16.5^{-1}+17.5^{-1}+18^{-1}+18^{-1}+20^{-1}+22^{-1}}=17.9037$$

3.  Wer von beiden hat Recht?

    Das harmonische Mittel wäre korrekt gewesen, wenn die Aufgabenstellung gelautet
    hätte: S beschließt, jeden Tag einen Kilometer Fahrrad zu fahren. Am ersten Tag
    schafft er den Kilometer mit einer Durchschnittsgeschwindigkeit von 15 km/h, am
    zweiten mit 16,5 km/h usw. …

4.  Um wie viel Prozent hat Sauerbrot seine Kilometerleistung im Schnitt gesteigert?

   **1. Leistungssteigerung:**
   
   $\frac{15}{100}=\frac{16.5}{x_1}$
   
   $x_1=\frac{100\cdot 16.5}{15}=110$
   
   Leistungssteigerung: $110\%-100\%=10\%$

   Wachstumsfaktor: $\frac{110}{100}=1.1$

   **2. Leistungssteigerung:**

   $\frac{16.5}{100}=\frac{17.5}{x_2}$

   $x_2=\frac{100\cdot 17.5}{16.5}=106.06$

   Leistungssteigerung: $106.05\%-100\%=6.05\%$

   Wachstumsfaktor: $\frac{106.06}{100}=1.0606$

   **3. Leistungssteigerung:**

   $\frac{17.5}{100}=\frac{18}{x_3}$

   $x_3=\frac{100\cdot 18}{17.5}=102.86$

   Leistungssteigerung: $102.86\%-100\%=2.86\%$

   Wachstumsfaktor: $\frac{102.86}{100}=1.0286$

   **4. Leistungssteigerung:**
 
    Leistungssteigerung: 0%

    Wachstumsfaktor: 1

    **5. Leistungssteigerung:**

    $\frac{18}{100}=\frac{20}{x_5}$

    $x_5=\frac{100\cdot 20}{18}=111.11$

    Leistungssteigerung: $111.11\%-100\%=11.11\%$

    Wachstumsfaktor: $\frac{111.11}{100}=1.1111$

    **6. Leistungssteigerung:**

    $\frac{20}{100}=\frac{22}{x_6}$

    $x_6=\frac{100\cdot 22}{20}=110$

    Leistungssteigerung: $110\%-100\%=10\%$

    Wachstumsfaktor: $\frac{110}{100}=1.1$
